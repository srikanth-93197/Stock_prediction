```python
import pandas as pd
from ntscraper import Nitter  
from datetime import datetime,timedelta
```

```python
scraper = Nitter()
```

```python
start_date = datetime(2023,1,1)
end_date = datetime(2023,1,2)
final_date = datetime(2024,2,15)
```

```python
final_tweets = []
```

```python
def TweetScraper(start_date,end_date,filters,terms):
    global final_tweets
    try:
        tweets = scraper.get_tweets(terms,mode='hashtag',since=start_date.strftime('%Y-%m-%d'),until=end_date.strftime('%Y-%m-%d'));
        # len = 0
        for tweet in tweets['tweets']:
            data = [tweet['user']['username'],tweet['date'],tweet['text'],tweet['stats']['likes']]
            final_tweets.append(data) 
    except Exception as e:
        print(f'{e}')
```

```python
filters = {}
terms = ['paytm']
```

```python
while end_date != final_date:
    # print(1000-counter)
    TweetScraper(start_date,end_date,filters,terms)
    start_date = start_date + timedelta(days=1)
    end_date = end_date + timedelta(days=1)
```

```python
data = pd.DataFrame(final_tweets,columns=['user','Date','text','likes'])
```
